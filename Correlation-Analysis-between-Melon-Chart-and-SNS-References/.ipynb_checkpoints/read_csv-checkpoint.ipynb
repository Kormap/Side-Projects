```python
melon_chart = tweet['melon_chart']
reference = tweet['count']
```

```python
import matplotlib.pyplot as plt
plt.plot(melon_chart, reference)
plt.show()
```

```python
sorted_reference = sorted(reference, reverse=True)
score = list()
melon_score = list()
for i in range(50):
    score.append(50 - sorted_reference.index(reference[i]))
    melon_score.append(51 - melon_chart[i])
```

```python
plt.figure(figsize=(10,8))
plt.plot(melon_score, score, color='r')
plt.xlabel('Melon Chart Ranking')
plt.ylabel('Reference Count Ranking')
plt.xticks()
plt.xticks(range(51), [50-x for x in range(-1,50)])
plt.yticks(range(51), [50-x for x in range(-1,50)])
plt.grid(True)
plt.show()
```

```python
import numpy as np
np.corrcoef(melon_score, score)
```

```python
import seaborn as sns
sns.pairplot(melon_score,score)
```
