```python
def get_rank(series, keyword):
    if len(series[series == keyword]) != 0:
        index = int(series[series == keyword].index[0].replace('Rank',''))
        time = series.values[0]
        return index
    else:
        return None
get_rank(read.iloc[-1], '진중권')
```

```python
def get_ranks(keyword):
    read = pd.read_csv("./data/data.csv", encoding="utf-8")
    ENGINE_NUM = 3
    minute = 1
    n_ranks = []
    d_ranks = []
    g_ranks = []
    for i in range(7):
        if get_rank(read.iloc[-1 - i * ENGINE_NUM * minute], keyword) != None:
            d_ranks.append([11 - get_rank(read.iloc[-1 - i * ENGINE_NUM * minute], keyword),
                           read.iloc[-1 - i * ENGINE_NUM * minute].values[0]])
        if get_rank(read.iloc[-2 - i * ENGINE_NUM * minute], keyword) != None:
            n_ranks.append([11 - get_rank(read.iloc[-2 - i * ENGINE_NUM * minute], keyword),
                           read.iloc[-2 - i * ENGINE_NUM * minute].values[0]])
        if get_rank(read.iloc[-3 - i * ENGINE_NUM * minute], keyword) != None:
            g_ranks.append([11 - get_rank(read.iloc[-3 - i * ENGINE_NUM * minute], keyword),
                           read.iloc[-3 - i * ENGINE_NUM * minute].values[0]])
    n_ranks.reverse()
    d_ranks.reverse()
    g_ranks.reverse()
    return n_ranks, d_ranks, g_ranks
n_ranks, d_ranks, g_ranks = get_ranks('도끼')
```
